```r
# install packages
library(tidyverse)
library(lubridate)
library(ggplot2)
setwd("~/Cyclistic Project")
getwd()
```

```r
# STEP 1: COLLECT DATA
# Upload Divvy datasets
library(readxl)
april_2020<- read_excel("202004-divvy-tripdata-xlsx.xlsx", 
                                          sheet = "tripdata-202004")
View(april_2020)
may_2020<- read_excel("202005-divvy-tripdata-xlsx.xlsx", 
                        sheet = "tripdata-202005")
View(may_2020)
jun_2020<- read_excel("202006-divvy-tripdata-xlsx.xlsx", 
                      sheet = "tripdata-202006")
View(jun_2020)
jul_2020<- read_excel("202007-divvy-tripdata-xlsx.xlsx", 
                      sheet = "tripdata-202007")
View(jul_2020)
aug_2020<- read_excel("202008-divvy-tripdata-xlsx.xlsx", 
                      sheet = "tripdata-202008")
View(aug_2020)
sep_2020<- read_excel("202009-divvy-tripdata-xlsx.xlsx", 
                      sheet = "tripdata-202009")
View(sep_2020)
oct_2020<- read_excel("202010-divvy-tripdata-xlsx.xlsx", 
                      sheet = "tripdata-202010")
View(oct_2020)
nov_2020<- read_excel("202011-divvy-tripdata-xlsx.xlsx", 
                      sheet = "tripdata-202011")
View(nov_2020)
des_2020<- read_excel("202012-divvy-tripdata-xlsx.xlsx", 
                      sheet = "tripdata-202012")
View(des_2020)
jan_2021<- read_excel("202101-divvy-tripdata-xlsx.xlsx", 
                      sheet = "tripdata-202101")
View(jan_2021)
feb_2021<- read_excel("202102-divvy-tripdata-xlsx.xlsx", 
                      sheet = "tripdata-202102")
View(feb_2021)
mar_2021<- read_excel("202103-divvy-tripdata-xlsx.xlsx", 
                      sheet = "tripdata-202103")
View(mar_2021)
```

```r
# STEP 2: WRANGLE DATA AND COMBINE INTO A SINGLE FILE
# Compare column names each of the files
colnames(april_2020)
colnames(may_2020)
colnames(jun_2020)
colnames(jul_2020)
colnames(aug_2020)
colnames(sep_2020)
colnames(oct_2020)
colnames(nov_2020)
colnames(des_2020)
colnames(jan_2021)
colnames(feb_2021)
colnames(mar_2021)
```

```r
# Inspect the dataframes and look for inconguencies
str(april_2020)
april_2020 <-  mutate(april_2020, start_lng = as.numeric(start_lng)
                   ,end_lat = as.numeric (end_lat)
                   ,end_lng = as.numeric(end_lng))

str(may_2020)
may_2020 <- mutate(may_2020, start_lng = as.numeric(start_lng)
                   ,end_lng = as.numeric(end_lng))

str(jun_2020)
str(jul_2020)
str(aug_2020)
aug_2020 <- mutate (aug_2020,start_lng = as.numeric(start_lng)
                    ,start_lat = as.numeric(start_lat)
                    ,end_lat = as.numeric (end_lat)
                    ,end_lng = as.numeric(end_lng))

str(sep_2020)
sep_2020 <- mutate (sep_2020,start_lng = as.numeric(start_lng)
                    ,start_lat = as.numeric(start_lat)
                    ,end_lat = as.numeric (end_lat)
                    ,end_lng = as.numeric(end_lng))

str(oct_2020)
oct_2020 <- mutate (oct_2020,start_lng = as.numeric(start_lng)
                    ,start_lat = as.numeric(start_lat)
                    ,end_lat = as.numeric (end_lat)
                    ,end_lng = as.numeric(end_lng))

str(nov_2020)
nov_2020 <- mutate (nov_2020,start_lng = as.numeric(start_lng)
                    ,start_lat = as.numeric(start_lat)
                    ,end_lat = as.numeric (end_lat)
                    ,end_lng = as.numeric(end_lng))

str(des_2020)
des_2020 <- mutate (des_2020,start_lng = as.numeric(start_lng)
                    ,start_lat = as.numeric(start_lat)
                    ,end_lat = as.numeric (end_lat)
                    ,end_lng = as.numeric(end_lng)
                    ,start_station_id = as.numeric (start_station_id)
                    ,end_station_id = as.numeric(end_station_id))

str(jan_2021)
jan_2021 <- mutate (jan_2021,end_lat = as.numeric (end_lat)
                    ,end_lng = as.numeric(end_lng)
                    ,start_station_id = as.numeric (start_station_id)
                    ,end_station_id = as.numeric(end_station_id))

# Fix NAs error in Jan_2021
jan_2021$start_station_id <- gsub(",", "", jan_2021$start_station_id) 
jan_2021$end_station_id <- gsub(",", "", jan_2021$end_station_id) 


str(feb_2021)
feb_2021 <- mutate (feb_2021,end_lat = as.numeric (end_lat)
                    ,end_lng = as.numeric(end_lng)
                    ,start_station_id = as.numeric (start_station_id)
                    ,end_station_id = as.numeric(end_station_id))

# Fix NAs error in Feb_2021
feb_2021$start_station_id <- gsub(",", "", feb_2021$start_station_id) 
feb_2021$end_station_id <- gsub(",", "", feb_2021$end_station_id) 


str(mar_2021)
mar_2021 <- mutate (mar_2021,end_lat = as.numeric (end_lat)
                            ,end_lng = as.numeric(end_lng)
                            ,start_station_id = as.numeric (start_station_id)
                            ,end_station_id = as.numeric(end_station_id))

# Fix NAs error in Mar_2021
mar_2021$start_station_id <- gsub(",", "", mar_2021$start_station_id) 
mar_2021$end_station_id <- gsub(",", "", mar_2021$end_station_id) 
```

```r
# Stack individual monthly data frames into one big data frame
all_trips <- rbind(jan_2021,feb_2021,mar_2021,april_2020,may_2020
                  ,jun_2020,jul_2020,aug_2020,sep_2020,oct_2020
                  ,nov_2020,des_2020)
```

```r
# Remove start_station_name,start_station_id, end_station_name,end_station_id
# as this data was dropped beginning in 2020
all_trips <- all_trips %>%  
  select(-c(start_station_name,start_station_id,end_station_name,end_station_id))
```

```r
# STEP 3: CLEAN UP AND ADD DATA TO PREPARE FOR ANALYSIS
# Inspect the new table that has been created
colnames(all_trips)  
nrow(all_trips)  
dim(all_trips)  
head(all_trips)
tail(all_trips)
str(all_trips) 
summary <- summary(all_trips)  
```

```r
# Begin by seeing how many observations fall under each usertype
member_summary <- table(all_trips$member_casual)
bicycle_type_summary <- table(all_trips$rideable_type)
```

```r
# Add columns that list the date, month, day, and year of each ride
all_trips$date <- as.Date(all_trips$started_at) 
all_trips$month <- format(as.Date(all_trips$date), "%m")
all_trips$day <- format(as.Date(all_trips$date), "%d")
all_trips$year <- format(as.Date(all_trips$date), "%Y")
all_trips$day_of_week <- format(as.Date(all_trips$date), "%A")
```

```r
# Add a "ride_length" calculation to all_trips (in seconds)
all_trips$ride_length <- difftime(all_trips$ended_at,all_trips$started_at)

# Inspect the structure of the columns
str(all_trips)
```

```r
# Convert "ride_length" from Factor to numeric so we can run calculations on the data
is.factor(all_trips$ride_length)
all_trips$ride_length <- as.numeric(as.character(all_trips$ride_length))
is.numeric(all_trips$ride_length)
```

```r
# Data inspection for NA,uncomplete case,bad data
list_unique <- lapply(all_trips, unique)    # List with unique values
list_unique                            # Print list

count_unique <- rapply(all_trips,  # Counts of unique values        
                       function(x) length(unique(x)))

colSums(is.na(all_trips))
rowSums(is.na(all_trips))
sum(!complete.cases(all_trips))
sum(is.na(all_trips$ride_length))
sum(all_trips$ride_length<0)
```

```r
# Remove "bad" data
# Remove ride_length<0
all_trips_v2 <- all_trips[!(all_trips$ride_length<0),]
sum(all_trips_v2$ride_length<0)
```

```r
# STEP 4: CONDUCT DESCRIPTIVE ANALYSIS
# Descriptive analysis on ride_length (all figures in seconds)
mean_ride_length <- mean(all_trips_v2$ride_length) #straight average (total ride length / rides)
median_ride_length <- median(all_trips_v2$ride_length) #midpoint number in the ascending array of ride lengths
max_ride_length <- max(all_trips_v2$ride_length) #longest ride
min_ride_length <- min(all_trips_v2$ride_length) #shortest ride
summary(all_trips_v2$ride_length)
```

```r
# Compare members and casual users
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = mean)
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = median)
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = max)
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = min)
```

```r
# See the average ride time by each day for members vs casual users
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual + all_trips_v2$day_of_week, FUN = mean)
```

```r
# Fix day of the week order.
all_trips_v2$day_of_week <- ordered(all_trips_v2$day_of_week, levels=c("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"))
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual + all_trips_v2$day_of_week, FUN = mean)
```

```r
# Analyze ridership data by type and weekday
all_trips_v2 %>% 
  mutate(weekday = wday(started_at, label = TRUE)) %>%  #creates weekday field using wday()
  group_by(member_casual, weekday) %>%  #groups by usertype and weekday
  summarise(number_of_rides = n()							#calculates the number of rides and average duration 
            ,average_duration = mean(ride_length)) %>% 		# calculates the average duration
  arrange(member_casual, weekday)								# sorts
```

```r
# Visualize the number of rides by rider type
all_trips_v2 %>% 
  mutate(weekday = wday(started_at, label = TRUE)) %>% 
  group_by(member_casual, weekday) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(member_casual, weekday)  %>% 
  ggplot(aes(x = weekday, y = number_of_rides, fill = member_casual)) +
  geom_col(position = "dodge")
```

```r
# Visualization for average duration
all_trips_v2 %>% 
  mutate(weekday = wday(started_at, label = TRUE)) %>% 
  group_by(member_casual, weekday) %>% 
  summarise(number_of_rides = n()
            ,average_duration = mean(ride_length)) %>% 
  arrange(member_casual, weekday)  %>% 
  ggplot(aes(x = weekday, y = average_duration, fill = member_casual)) +
  geom_col(position = "dodge")
```

```r
# STEP 5: EXPORT SUMMARY FILE FOR FURTHER ANALYSIS
counts <- aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual + all_trips_v2$day_of_week, FUN = mean)
write.csv(counts, file = "C:\\Users\\PC Dekstop\\Documents\\Cyclistic Project\\avg_ride_length.csv")
write.csv(all_trips_v2, file = "C:\\Users\\PC Dekstop\\Documents\\Cyclistic Project\\all_trips.csv")
```
